**O problema: O que queremos Predizer?**

O Delorean (DMC-12) foi o único carro fabricado pela empresa automobilística estadunidense DeLorean Motor Company (DMC).Tendo sua linha de produção ativa apenas entre 1981 e 1982, podemos aferir que diante a história de outros modelos produzidos por outras montadoras, o seu período de existência foi breve. Porém, apesar da pequena quantidade de modelos produzidos nesse breve período de tempo, esse modelo foi eternizado pela trilogia de filmes "De Volta para o Futuro".

Em sua versão de fábrica o delorean desenvolve uma potência de 130 HP (132 CV) a 5.500 rpm e um torque de 207 N/m a 2750 rpm. Entretanto, a versão que encontramos no filme passa por constantes modificações ao longo dos anos. No primeiro filme da sequência, por exemplo, na ocasião da primeira viagem o carro parece ser uma espécie de híbrido entre motorização a gasolina e plutônio. Já na segunda viagem, é a energia gerada com o raio que vem auxiliar na performance do carro/máquina do tempo. E, ao final do filme, o Dr. Emmett Brown aparece com uma nova versão alimentada por matéria orgânica proveniente do lixo. Será que essas alterações no modelo geraram alguma alteração em sua potência original? É isso que o nosso modelo de Machine Learning irá nos ajudar a predizer.

**A tarefa: Como iremos fazer a predição?**

Após uma busca por Data Sets em repositórios de livre acesso, encontramos no repositório de Machine Learning da UCI um conjunto de dados intitulados "Auto MPG Data Set". Esse foi um data set usado em 1983 pela Associação Expositória de Estatísticas Americanas e guardado pela Universidade de Carnegie Mellhon, dentre os atributos, possuímos os seguintes marcadores: 'mpg'; 'cylinders', 'displacement'; 'horsepower'; 'weight'; 'acceleration'; 'model year'; 'origin'; e 'car name'. Nosso objetivo é calcular a potência do veículo em cavalos (horsepower), a partir da aceleração fornecida pelo modelo. Assim, utilizaremos a tarefa de regressão linear utilizando apenas os atributos 'horsepower'' e "acceleration" do dataset para treinar, testar e otimizar o nosso modelo. A escolha desses atributos se deu por sabermos que no mundo real a aceleração de um modelo está diretamente relacionada com a sua potência. Assim, estaríamos criando um modelo com alguma pertinência aplicabilidade real, embora entendamos que tenham outras variáveis que possam interferir nas relações entre esses dois atributos. Além disso, outro fator importante para a escolha deste dataset e dos atributos utilizados foi a possibilidade de coleta de dados a partir de informações do filme. Isso porque, embora não seja informada nenhuma especificação do delorean de forma direta no filme. Podemos calcular a aceleração do modelo depois de cada alteração promovida pelo Dr. Brown a partir da análise do tempo que o delorean levou para atingir a velocidade máxima para viagem no tempo (140 km/h) em cada momento do filme.

**Link do Data Set no Repositório UCI:** https://archive.ics.uci.edu/ml/datasets/Auto+MPG

# Importação de pacotes

```
import pandas as pd
import io
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

# Apresentação e carregamento dos dados

Explique onde os dados podem ser acessados, do que se tratam, quais são as informações providas, a razão de escolherem esse conjunto.

```
url = 'https://raw.githubusercontent.com/lucascastrop/Datasets/main/auto-mpg-pronto.csv'
data = pd.read_csv(url)
data.head()
```

# Análise Exploratória dos dados


Veja quais são as colunas existentes, se tem dado faltante, quais são os tipos dos dados etc e descreva a conclusão de cada passo.

```
data_numeric = data[['acceleration', 'horsepower', 'weight']]

data_numeric = data_numeric.dropna()
data_numeric.tail(5)

aceleracao_peso =  data_numeric[['acceleration', 'weight']]
cv = data_numeric[['horsepower']]


print(aceleracao_peso.head())
print(cv.head())


```

# Definição de x e y e separação dos subconjuntos de treinamento e de teste

Explique porque essas separações precisa ser realizadas e mostre como fazer.

```
x= aceleracao_peso.values
y = cv.values

x_train, x_test, y_train, y_test = train_test_split(x, y, 
                                                    test_size = 0.3, 
                                                    random_state = 42)
```

## [Modelo escolhido - Regressão Linear]

Explique a razão de escolha desse modelo e como ele funciona.

### Treinamento do modelo

```
model = LinearRegression().fit(x_train, y_train)
```

### Avaliação

Quais foram os resultados alcançados? O que eles significam? Pode ter ocorrido overfitting?

```
y_pred = model.predict(x_test)
mean_squared_error(y_test, y_pred)
```

```
mean_absolute_error(y_test, y_pred)
```

```
r2_score(y_test, y_pred)
```
